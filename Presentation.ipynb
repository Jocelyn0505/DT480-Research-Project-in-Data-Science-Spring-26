```python
import pandas as pd 
import kagglehub
import os

path = kagglehub.dataset_download("yaninapetrongari/crime-california-work-reboot")

print("Path to dataset files:", path)

# 2. Automatically find the CSV file in the path
files = [f for f in os.listdir(path) if f.endswith('.csv')]
if not files:
    raise FileNotFoundError("No CSV file found in the downloaded folder!")

csv_path = os.path.join(path, files[0])
print(f"Loading: {csv_path}")

# 3. Load the data
df = pd.read_csv(csv_path)

# List of keywords to identify columns for removal
keywords = [
    "Month", "DR_NO", "Mocodes", "Vict Descent", "Premis Cd", 
    "Weapon Used Cd", "Weapon Desc", "Status", "Status Desc", 
    "AREA NAME", "Part 1-2", "Date Rptd", "TIME OCC2", "LOCATION"]

# Drop all columns that contain any of the keywords
df = df.drop(columns=[col for col in df.columns if any(word in col for word in keywords)])

df
```

```python
df["DATE OCC"] = pd.to_datetime(df["DATE OCC"])
filtered_df = df[
    (df["DATE OCC"] >= "2020-01-01") &
    (df["DATE OCC"] <= "2023-12-31")
]
filtered_df
```

```python
filtered_df = filtered_df.sort_values(by="DATE OCC", ascending=False)
filtered_df
```

```python
import geopandas as gpd
from shapely.geometry import Point

# 1. Convert your DataFrame to a GeoDataFrame
# We create a 'geometry' column from LAT and LON
geometry = [Point(xy) for xy in zip(filtered_df['LON'], filtered_df['LAT'])]
gdf_crime = gpd.GeoDataFrame(filtered_df, geometry=geometry)

# Set the coordinate system to WGS84 (standard Lat/Lon)
gdf_crime.set_crs(epsg=4326, inplace=True)

# 2. Load a map of US Counties 
# This is a low-res map included with geopandas or available via URL
url = "https://www2.census.gov/geo/tiger/GENZ2020/shp/cb_2020_us_county_5m.zip"
counties = gpd.read_file(url)

# Filter for California (State FP '06') to keep it fast
ca_counties = counties[counties['STATEFP'] == '06'].to_crs(epsg=4326)

# 3. The "Spatial Join" - This is the magic part
# It checks which county polygon contains each crime point
joined = gpd.sjoin(gdf_crime, ca_counties[['NAME', 'geometry']], how='left', predicate='intersects')

# 4. Rename and cleanup
filtered_df['county'] = joined['NAME']
filtered_df
```

```python
filtered_df["county"].value_counts()
```

```python
keywords = ["LAT", "LON"]

filtered_df = filtered_df.drop(columns=[col for col in filtered_df.columns if any(word in col for word in keywords)])

filtered_df
```
